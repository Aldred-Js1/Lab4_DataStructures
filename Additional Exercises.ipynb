```python
SEED_NUM = 3
STUDENT_MAJOR = "BSECE"
FAVORITE_ARTIST = "HIROHIKO ARAKI"
LOCAL_HAZARD = "FLOOD"
CONTROL_NUM = max(3, SEED_NUM)
```

```python
# EXERCISE 1

from collections import defaultdict, Counter

# 1. Initialize variables (Ensure these were run in your initialization cell)
# SEED_NUM = 3
# STUDENT_MAJOR = "BSECE"
# FAVORITE_ARTIST = "HIROHIKO ARAKI"
# LOCAL_HAZARD = "FLOOD"
# CONTROL_NUM = 3

# 2. Initialize the simulated stream of incoming applications
stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

# 3. Data Expansion: Add exactly (CONTROL_NUM + 3) = 6 custom application tuples
custom_applications = [
    ("TUP Manila", "BSECE"), # Adding more BSECE to Manila
    ("TUP Manila", "BSIT"),
    ("TUP Cavite", "BSECE"),
    ("TUP Taguig", "BSIT"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", "BSIT")
]
stream.extend(custom_applications)

# 4. Algorithm Phase 1 (Aggregation): Group programs by Campus key
campus_map = defaultdict(list)
for campus, program in stream:
    campus_map[campus].append(program)

# 5. Algorithm Phase 2 (Frequency Analysis): Map frequencies using Counter
campus_frequencies = {}
for campus, programs in campus_map.items():
    campus_frequencies[campus] = Counter(programs)

# 6. Extract the top requested program for Manila campus
manila_top = campus_frequencies["TUP Manila"].most_common(1)

# 7. Print Outputs for the Assessment Table
print("--- EXERCISE 1: ADMISSIONS SUMMARY ---")
print(f"Total Stream Length: {len(stream)}")
for campus, counts in campus_frequencies.items():
    print(f"{campus.upper()}: {dict(counts)}")

if manila_top:
    prog, freq = manila_top[0]
    print(f"\nMANILA ANALYSIS:")
    print(f"Top Program: {prog}")
    print(f"Frequency: {freq}")
```

```python
# EXERCISE 2

# 1. Initialize the official Festival Lineup as a Set
# Using your specific FAVORITE_ARTIST and CONTROL_NUM
festival = {
    "Ben&Ben", "SB19", "Bini", "Eraserheads", 
    FAVORITE_ARTIST, "Zild", f"Indie Artist {CONTROL_NUM}"
}

# 2. Initialize two User Profiles as Sets
user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {
    "SB19", "Eraserheads", "Zild", 
    f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"
}

# 3. Data Expansion: Use .update() to add exactly CONTROL_NUM (3) new artists
# You can choose any 3 Filipino artists here
new_artists = {"Lola Amour", "Dilaw", "Cup of Joe"}
festival.update(new_artists)

# 4. Algorithm Phase 1: Jaccard Similarity Calculation
# Formula: J(A, B) = |Intersection| / |Union| * 100
def get_jaccard_index(user_set, festival_set):
    intersection = user_set.intersection(festival_set)
    union = user_set.union(festival_set)
    # Using float division for accuracy
    return (len(intersection) / len(union)) * 100

# Calculate scores for User A and User B
score_a = get_jaccard_index(user_a, festival)
score_b = get_jaccard_index(user_b, festival)

# 5. Algorithm Phase 2: Set Difference for "Dealbreakers"
# Artists User A likes that are NOT performing at the festival
dealbreakers_a = user_a.difference(festival)

# 6. Print Outputs for the Assessment Table
print("== EXERCISE 2: JACCARD SIMILARITY REPORT ==\n")
print(f"Total Artists in Expanded Festival: {len(festival)}")
print(f"User A Intersection Length: {len(user_a.intersection(festival))}")
print(f"User A Union Length: {len(user_a.union(festival))}")
print(f"User A Jaccard Similarity: {score_a:.2f}%")
print(f"User B Jaccard Similarity: {score_b:.2f}%")
print(f"User A Dealbreaker Artists: {dealbreakers_a}")
```

```python
# EXERCISE 3

from collections import deque, Counter

# 1. Initialize a deque with a strict memory limit: maxlen = 5
alert_buffer = deque(maxlen=5)

# 2. Simulate the initial burst of incoming alerts
# Starting at Timestamp = CONTROL_NUM (3)
burst = [
    (3, "WEATHER"),
    (4, "TRAFFIC"),
    (5, "WEATHER"),
    (6, "WEATHER"),
    (7, "FLOOD"),   # Your LOCAL_HAZARD
    (8, "WEATHER"),
    (9, "TRAFFIC")
]

# 3. Data Expansion: Add exactly CONTROL_NUM + 2 (5) custom alerts
# Continuing the timestamps sequentially from 10 to 14
custom_alerts = [
    (10, "FIRE"),
    (11, "FLOOD"),
    (12, "TRAFFIC"),
    (13, "FLOOD"),
    (14, "MEDICAL")
]
burst.extend(custom_alerts)

# 4. Algorithm: The Sliding Window Filter
print("== EMERGENCY NOTIFICATION SYSTEM LOG ==\n")
for alert in burst:
    # Extract just the categories currently sitting inside the deque
    current_categories = [item[1] for item in alert_buffer]
    counts = Counter(current_categories)
    
    incoming_category = alert[1]
    
    # Spam Rule: Reject if category exists 2 or more times in the current window
    if counts[incoming_category] >= 2:
        print(f"REJECTED: {alert} - Category '{incoming_category}' limit reached.")
    else:
        alert_buffer.append(alert)
        print(f"ACCEPTED: {alert}")

# 5. Final Calculation: Sum of Timestamps in the deque
final_timestamp_sum = sum(alert[0] for alert in alert_buffer)

# 6. Outputs for your Assessment Table
print("\n== ASSESSMENT DATA OUTPUT ==")
print(f"Final Deque State: {list(alert_buffer)}")
print(f"Final Sum of Timestamps: {final_timestamp_sum}")
```
